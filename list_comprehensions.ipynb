```python
colors = ['black', 'white']
sizes = ['S', 'M', 'L']
pairs=[(color,size) for color in colors for size in sizes]
pairs
```

```python
def listcomp():
    colors = [i for i in range(1000)]
    sizes = [i for i in range(1000)]
    pairs=[(size,color) for size in sizes for color in colors]
    return pairs
```

```python
def genexp():
    colors = [i for i in range(1000)]
    sizes = [i for i in range(1000)]
    pairs=((size,color) for size in sizes for color in colors)
    return pairs
```

```python
%timeit -n 10 -r 5 listcomp()
%timeit -n 10 -r 5 genexp()
```

```python
# Генераторное выражение не использует память
colors = ['black', 'white'] 
sizes = ['S', 'M', 'L']
for tshirt in (f'{c} {s}' for c in colors for s in sizes):
    print(tshirt)
```
