```python
import csv
import json
import pandas as pd

with open('C:/Users/fedor/Downloads/RU.txt', 'r', encoding='utf-8') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    data = []
    for row in reader:
        if row and len(row) > 10:
            data.append({
                "name": row[1],
                "latitude": float(row[4]),
                "longitude": float(row[5]),
                "population": int(row[14]) if row[14].isdigit() else 0
            })

with open('russia_cities.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
cities = []
for city in data:
    cities.append(city['name'])

latitude=[]
longitude=[]
for city in data:
    latitude.append(city['latitude'])
    longitude.append(city['longitude'])

df = pd.DataFrame(list(zip(cities,latitude,longitude)), columns=['city', 'latitude','longitude'])
df.to_csv('C:/Users/fedor/Downloads/all_rus_cities.csv', sep='|', encoding='utf-8-sig') 
```
