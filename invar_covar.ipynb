```python
# Пример инвариантности
# Инвариантность требует точного совпдаения типов, родительские и дочерние классы также вызовут ошибку
from typing import TypeVar, Generic
class Beverage: 
    """Любой напиток."""

class Juice(Beverage):
    """Любой фруктовый сок."""

class OrangeJuice(Juice):
    """Восхитительный сок бразильских апельсинов."""

T = TypeVar('T')

class BeverageDispenser(Generic[T]):
    """Автомат, параметризованный типом напитка."""
    def __init__(self, beverage: T) -> None:
        self.beverage = beverage

    def dispense(self) -> T:
        return self.beverage
    
def install(dispenser: BeverageDispenser[Juice]) -> None:
        """Установить автомат для розлива фруктовых соков."""
```

```python
install(BeverageDispenser(Juice())) # средство статической проверки не вызовет ошибки
```

```python
install(BeverageDispenser(Beverage())) # средство статической проверки вызовет ошибку, так как явно указан класс Juice, его родительский класс тоже вызывает ошибку
```

```python
install(BeverageDispenser(OrangeJuice())) # средство статической проверки вызовет ошибку, так как явно указан класс Juice, его дочерний класс тоже вызывает ошибку
```

```python
# Пример ковариантности
# Проходить проверку будет Juice2 и его подклассы, а родитель нет
from typing import TypeVar, Generic
class Beverage2: 
    """Любой напиток."""

class Juice2(Beverage2):
    """Любой фруктовый сок."""

class OrangeJuice2(Juice2):
    """Восхитительный сок бразильских апельсинов."""

T_co = TypeVar('T_co', covariant=True)

class BeverageDispenser2(Generic[T]):
    """Автомат, параметризованный типом напитка."""
    def __init__(self, beverage: T_co) -> None:
        self.beverage = beverage

    def dispense(self) -> T_co:
        return self.beverage
    
def install(dispenser: BeverageDispenser2[Juice2]) -> None:
        """Установить автомат для розлива фруктовых соков."""
```

```python
install(BeverageDispenser2(Beverage2())) # средство статической проверки вызовет ошибку
```

```python
install(BeverageDispenser2(Juice2())) # средство статической проверки не вызовет ошибки
```

```python
install(BeverageDispenser2(OrangeJuice2())) # средство статической проверки тоже не вызовет ошибки, так как это дочерний класс и мы включили ковариантность
```

```python
# Пример контравариантности
# Контравариантность меняет направление отношения наследования
# В результате подклассы Biodegradable проверку не пройдут, а родители пройдут.
from typing import TypeVar, Generic
class Refuse: 
    """Любые отходы."""

class Biodegradable(Refuse):
    """Биоразлагаемые отходы."""

class Compostable(Biodegradable):
    """Компостируемые отходы."""

T_contra = TypeVar('T_contra', contravariant=True)

class TrashCan(Generic[T_contra]):

    def __init__(self, refuse: T_contra)-> None:
        self.refuse = None

    def put(self, refuse: T_contra) -> None:
        """Хранить отходы, пока не выгружены."""

def deploy(trash_can: TrashCan[Biodegradable]):
        """Установить урну для биоразлагаемых отходов."""
```

```python
deploy(TrashCan(Biodegradable())) # средство статической проверки не вызовет ошибки
```

```python
deploy(TrashCan(Refuse())) # средство статической проверки не вызовет ошибки
```

```python
deploy(TrashCan(Compostable())) # средство статической проверки вызовет ошибку
```
