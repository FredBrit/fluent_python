```python
import csv
import unicodedata

def clean_text(s):
    if not isinstance(s, str):
        return s
    # Удаляем непечатаемые символы (кроме пробелов, табов, переносов)
    return ''.join(c for c in s if c.isprintable() or c in ' \t\n\r')

input_file = 'C:\Users\fedor\Downloads\BI-академия\couriers.csv'
output_file = '"C:\Users\fedor\Downloads\couriers_cleaned.csv'

with open(input_file, 'r', encoding='utf-8') as fin, \
     open(output_file, 'w', encoding='utf-8', newline='') as fout:
    
    reader = csv.reader(fin)
    writer = csv.writer(fout)
    
    for row in reader:
        cleaned_row = [clean_text(cell) for cell in row]
        writer.writerow(cleaned_row)

print("✅ CSV очищен от невидимых символов.")
```
