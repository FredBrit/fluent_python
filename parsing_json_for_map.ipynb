```python
import json
import pandas as pd

with open('C:/Users/Fedor/Downloads/jsonviewer.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
data

cities = []
for city in data:
    cities.append(city['cities'])
cities

names=[]
coordinates=[]
for el in cities:
    for el in el:
        names.append(el['name'])
        coordinates.append(el['coordinates'])
names,coordinates

df = pd.DataFrame(list(zip(names,coordinates)), columns=['city', 'coordinates',])
df1=df['coordinates'].str.split(',', expand=True)
df_total=pd.concat([df,df1],axis=1)
df_total=df_total.rename(columns={0:'geo_lan',1:'geo_lat'})
df_total

df_total.to_csv('Для карты.csv', sep='|', encoding='utf-8-sig')
```
