```python
from dataclasses import dataclass, field

# Обычный класс — много кода
class ClubMemberManual:
    def __init__(self, name: str, guests: list = None, athlete: bool = False):
        self.name = name
        self.guests = guests if guests is not None else []
        self.athlete = athlete
    
    def __repr__(self):
        return f"ClubMemberManual(name={self.name!r}, guests={self.guests!r})"
    
    def __eq__(self, other):
        if not isinstance(other, ClubMemberManual):
            return NotImplemented
        return (self.name, self.guests, self.athlete) == (other.name, other.guests, other.athlete)

# Dataclass — компактно
@dataclass
class ClubMember:
    name: str
    guests: list = field(default_factory=list)
    athlete: bool = field(default=False, repr=False)
```
