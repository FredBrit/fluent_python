```python
# Пример реализации основных специальных методов
from array import array
import math
class Vector2d:

    # для использования позиционных аргументов
    __match_args__ = ('x', 'y')

    typecode = 'd'
    def __init__(self, x, y):
        self.__x = float(x)
        self.__y = float(y)

    @property
    def x(self):
        return self.__x
    
    @property
    def y(self):
        return self.__y
    
    def __hash__(self):
        return hash((self.x, self.y))

    def __iter__(self):
        return (i for i in (self.x, self.y))
    
    def __repr__(self):
        class_name = type(self).__name__
        return '{}({!r}, {!r})'.format(class_name, *self)
    
    def __str__(self):
        return str(tuple(self))
    
    def __bytes__(self):
        return (bytes([ord(self.typecode)]) +
                bytes(array(self.typecode, self)))
    
    def __eq__(self, other):
        return tuple(self) == tuple(other)
    
    def __abs__(self):
        return math.hypot(self.x, self.y)
    
    def __bool__(self):
        return bool(abs(self))
    
    def __format__(self, fmt_spec=''):
        if fmt_spec.endswith('p'):
            fmt_spec = fmt_spec[:-1]
            coords = (abs(self), self.angle())
            outer_fmt = '<{}, {}>'
        else:
            coords = self
            outer_fmt = '({}, {})'
        components = (format(c, fmt_spec) for c in coords)
        return outer_fmt.format(*components)

    @classmethod
    def frombytes(cls, octets):
        typecode = chr(octets[0])
        memv = memoryview(octets[1:]).cast(typecode)
        return cls(*memv)
    
def positional_pattern_demo(v: Vector2d) -> None:
    match v:
        case Vector2d(0, 0):
            print(f'{v!r} is null')
        case Vector2d(0):
            print(f'{v!r} is vertical')
        case Vector2d(_, 0):
            print(f'{v!r} is horizontal')
        case Vector2d(x, y) if x==y:
            print(f'{v!r} is diagonal')
        case _:
            print(f'{v!r} is awesome')    
```

```python
class Pixel:
    __slots__ = ('x','y') # может быть список


class OpenPixel(Pixel):
    pass

class ColorPixel(Pixel):
    __slots__ = ('color',)
```

```python
# Атрибуты больше не хранятся в __dict__ и самого __dict__ больше не существует
p=Pixel()
p.__dict__
```

```python
# Присваивание значения указанному в __slots__ атрибуту
p.x=10
p.x
```

```python
# Создание несуществующего в slots атрибута невозможно
p.z=8
p.z
```

```python
# Несмотря на то, что OpenPixel наследуется от Pixel, в нем существует __dict__
op=OpenPixel()
op.__dict__
```

```python
# Присваивание существующму в slots атрибуту-возможно
op.x=8
op.x
```

```python
# Присваивание НЕсуществующму в slots атрибуту- ТОЖЕ возможно
# в __dict__ создается новый атрибут
op.z=10
op.__dict__
```

```python
# При указании в наследуемом классе slots - dict уничтожается
cp=ColorPixel()
cp.__dict__
```

```python
# Присваивание существующму в slots класса родителя атрибуту-возможно
cp.x=10
cp.x
```

```python
# Присваивание существующму в slots атрибуту-возможно
cp.color='red'
cp.color
```

```python
# Присваивание НЕсуществующму в slots атрибуту - НЕвозможно
cp.z=10
cp.z
```
