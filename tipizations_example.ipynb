```python
# Duck Typing (Утиная типизация)
# Наследование одного и тоже метода, пусть и с разный по смыслу реализацией
def quack(duck):
    duck.quack()  # Если объект "крякает", он утка!

class Duck:
    def quack(self): print("Quack!")

class Dog:
    def quack(self): print("Woof!")  # Собака тоже может "крякать"

quack(Duck())  # Quack!
quack(Dog())   # Woof! — работает, хоть и не утка
```

```python
# EAFP (Easier to Ask for Forgiveness than Permission)
# Лучше попросить прощения, чем разрешения
def get_first_item(items):
    try:
        return items[0]  # Просто пробуем
    except (IndexError, TypeError):
        return None      # Обрабатываем ошибку

print(get_first_item([1, 2, 3]))  # 1
print(get_first_item((1, 2, 3)))  # 1
print(get_first_item([]))         # None
print(get_first_item(None))       # None
```

```python
# LBYL (Look Before You Leap)
# Смотри, прежде чем прыгать
def get_first_item(items):
    if items and len(items) > 0:  # Проверяем заранее
        return items[0]
    return None

print(get_first_item([1, 2, 3]))  # 1
print(get_first_item([]))         # None
print(get_first_item(None))       # None
```

```python
# Строгая типизация через isinstance()
def process_number(x):
    if isinstance(x, (int, float)):  # Строгая проверка типа
        return x * 2
    raise TypeError("Ожидается число")

print(process_number(5))    # 10
print(process_number("5"))  # TypeError!
```

```python
# Протоколы (Structural Subtyping)
# Проверяем по наличию метода
from typing import Protocol

# Шаг 1: Определяем протокол
class Drawable(Protocol):
    def draw(self) -> None: ...

# Шаг 2: Функция принимает любой объект, соответствующий протоколу
def render(obj: Drawable) -> None:
    obj.draw()

# Шаг 3: Создаём класс, который НЕ наследуется от Drawable, но имеет метод draw
class Circle:
    def draw(self): 
        print("Рисую круг")

# Шаг 4: Используем
render(Circle())  # Работает!
```

```python
# Аннотации типов (Type Hints)
# Ошибка возникнет только при прогоне скрипта через средство статической проверки (mypy, pytest)
def greet(name: str) -> str:
    return f"Hello, {name}!"

message: str = greet("Alice")
message2: str = greet(123)
print(message)  # Hello, Alice!
print(message2)  # Hello, 123!
```
